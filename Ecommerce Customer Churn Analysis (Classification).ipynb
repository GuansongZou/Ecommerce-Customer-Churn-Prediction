### DATA MODELING

##### OneHotEncoder (encoding categorical features)

```python
#Link: https://datagy.io/sklearn-one-hot-encode/
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
```

```python
ohe.fit_transform(df[["PreferredLoginDevice"]])
ohe.fit_transform(df[["PreferredPaymentMode"]])
ohe.fit_transform(df[["Gender"]])
ohe.fit_transform(df[["PreferedOrderCat"]])
ohe.fit_transform(df[["MaritalStatus"]])
```

##### Make_column_transformer (adding columns)

```python
from sklearn.compose import make_column_transformer

transformer = make_column_transformer(
    (OneHotEncoder(), ['PreferredLoginDevice',"PreferredPaymentMode","Gender","PreferedOrderCat","MaritalStatus"]),
    remainder='passthrough')

transformed=transformer.fit_transform(df)

df = pd.DataFrame(transformed, columns=transformer.get_feature_names())
```
